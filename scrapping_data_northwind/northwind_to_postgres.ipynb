```python
import requests
import psycopg2
```

```python
# URL raw GitHub
SQL_URL = "https://raw.githubusercontent.com/pthom/northwind_psql/master/northwind.sql"
```

```python
# Koneksi PostgreSQL
DB_CONFIG = {
    "host": "postgres",
    "port": 5432,
    "dbname": "batch_db",
    "user": "airflow",
    "password": "airflow"
}
```

```python
# 1) Download script SQL
resp = requests.get(SQL_URL)
resp.raise_for_status()
sql_script = resp.text
```

```python
# 2) Koneksi ke DB
conn = psycopg2.connect(**DB_CONFIG)
conn.autocommit = True
cur = conn.cursor()
```

```python
print("Extract success, bytes:", len(sql_script))
```

```python
try:
    # 3) Pastikan schema northwind aktif
    cur.execute("""
        CREATE SCHEMA IF NOT EXISTS northwind;
        SET search_path TO northwind;
    """)

    # 4) Jalankan seluruh SQL
    cur.execute(sql_script)

    print("✅ Berhasil import ke schema northwind!")

except Exception as e:
    print("❌ Error:", e)

finally:
    cur.close()
    conn.close()
```
